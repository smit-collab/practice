## LINE CHART

#### Total profit of all months and show line plot with the following Style properties

```python
month = df['month_number'].tolist()
profit = df['total_profit'].tolist()
plt.plot(month,profit,label='profit of months',color='b',marker='o',markerfacecolor='k'
,linestyle='--',linewidth=3)

plt.xlabel('month')
plt.xticks(month)
plt.ylabel('profit')
plt.yticks([100000,200000,300000,400000,500000])
plt.show()
```

### Reading all product sales data and show it  using a multiline plot

```python

month = df['month_number'].tolist()
facecream = df['facecream'].tolist()
facewash = df['facewash'].tolist()
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
# moisturizer = df['moisturizer'].tolist()


plt.plot(month,facecream,label='Face cream sales',marker='o',markerfacecolor='k',linestyle='--',linewidth=3)
plt.plot(month,facewash,label='facewash sales',marker='o',markerfacecolor='k',linestyle='--',linewidth=3)
plt.plot(month,toothpaste,label='toothpaste  sales',marker='o',markerfacecolor='k',linestyle='--',linewidth=3)
plt.plot(month,bathingsoap,label='bathingsoap sales',marker='o',markerfacecolor='k',linestyle='--',linewidth=3)
plt.plot(month,shampoo,label='shampoo sales',marker='o',markerfacecolor='k',linestyle='--',linewidth=3)


plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(month)
# plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
# plt.figure(figsize=(5,5))
plt.show()
```

### toothpaste sales data of each month and show it using a scatter plot

```python
month = df['month_number'].tolist()
toothpaste = df['toothpaste'].tolist()

plt.scatter(month,toothpaste,label = 'toothpaste sales data')
plt.xlabel('Month')
plt.xticks(month)
plt.title('No of units sold')
plt.grid(True,linewidth= 1, linestyle="--")
plt.ylabel('Toothpaste sales')
plt.show()
```

```python
face_cream = df['facecream'].tolist()
face_wash = df['facewash'].tolist()

plt.scatter(face_cream,face_wash,label = 'hmm')
plt.xlabel('face_cream')
# plt.xticks(month)
# plt.title('No of units sold')
plt.grid(True,linewidth= 1, linestyle="--")
plt.ylabel('face_wash')
plt.show()
```

Not that strong relationship between face wash and face cream

### face cream and facewash product sales data and show it using the bar chart

```python
plt.bar([a-0.25 for a in month], face_cream,width=0.25,label='face cream sales data',align='edge')
plt.bar([a+0.25 for a in month], face_wash,width=-0.25,label='face wash sales data',align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(month)
plt.title('Facewash and facecream sales data')
plt.show()
```

###  total sale data for each product and show it using a Pie chart

```python
labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
sales_data = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(sales_data, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
